```python
num1 = int(input("Enter 1st number"))
num2 = int(input("Enter 2nd number"))
num3 = int(input("Enter 3rd number"))

#using max() function
print("The maximum number among three numbers:", max(num1,num2,num3))

# using if-elif-else condition
if num1 > num2 and num1 > num3:
    largest = num1
    print("The largest number is", largest)
elif num2 > num1 and num2 > num3:
    largest = num2
    print("The largest number is", largest)
else:
    largest = num3
    print("The largest number is", largest)
```

```python
# input product code & order amount
product_code = int(input("Enter the product code (1 for Battery, 2 for Key, 3 for Charging): "))
order_amount = float(input("Enter the order amount in Rs: "))

# initialize discount and net_amount variables
discount = 0
net_amount = 0

# Check the product code & apply discount accordingly
if product_code == 1:  # Battery-based Toys
    if order_amount > 1000:
        discount = 0.10
elif product_code == 2:  # Key-based Toys
    if order_amount > 100:
        discount = 0.05
elif product_code == 3:  # Charging-based Toys
    if order_amount > 500:
        discount = 0.10

# Calculate the net amount after applying the discount
net_amount = order_amount * discount

# print the net amount the customer is required to pay
print(f"Net amount after discount: Rs. {net_amount:2f}")    
```

```python
# Ask user to enter the past reading & current meter reading.
# Find the units consumed.
# Compute the bill according to given matrix.
```

```python
# Step 1: Ask the user to enter past and current meter reading.
past_reading = int(input("Enter the past meter reading"))
current_reading = int(input("Enter the current meter reading"))

# Step 2: Calculate units consumed
units_consumed = current_reading - past_reading

# Step 3: Compute the bill according to the given matrix
if units_consumed <= 100 :
    bill = units_consumed * 0.50
elif units_consumed <= 200 :
    bill = 50 + (units_consumed) * 1.00
elif units_consumed <= 300 :
    bill = 150 + (units_consumed) * 1.50
else:
    bill = 300 + (units_consumed) * 2.00
    
# Print the electricity bill
print("Unit Consumed", units_consumed)
print("Electricity Bill", bill)
```
