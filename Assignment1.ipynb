### Q11. Mention a few built-in functions in Python?

```python
# max()
l = [10,20,50,40,70,100]
print("max in list:",max(l))
print("min in list:",min(l))
print("length of list:",len(l))
print("sum of all list elements:",sum(l))
```

```python
# int()
percentage = 89.89
percentage = int(percentage)
print(percentage)
percentage = 89.89
# round()
percentage = round(percentage)
print(percentage)
# bool()
percentage = bool(percentage)
print(percentage)
```
